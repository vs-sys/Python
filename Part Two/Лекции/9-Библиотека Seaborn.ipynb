<h2 style="text-align:center">Категориальные графики</h2>

<h3 style="text-align:center">Функция <strong>catplot()</strong></h3>

<div style="text-indent:30px; text-align:justify">Рассмотренные ранее диаграммы используются для анализа исследуемых непрерывных переменных. Эти графики не подходят, когда исследуемая переменная является категориальной. </div>
<div style="text-indent:30px; text-align:justify">Категориальные графики необходимы, если значения переменной могут быть разделены на категории (подмножества) и показывают распределения внутри каждой категории.</div>
<div style="text-indent:30px; text-align:justify">По функциональным возможностям и назначению функция <strong>catplot()</strong> подобна ранее рассмотренной функции <strong>relplot()</strong>. Главная особенность этой функции  заключается в том, что она предоставляет общий интерфейс для всех функций из группы визуализации категориальных данных. Тип функции задается параметром <strong>kind</strong>.</div>

<h4 style="text-align:center">Диаграмма рассеяния, состоящая из одной категориальной переменной</h4>

```python
sns.catplot(data=df, 
            x="species",     # Вид
            y="body_mass_g", # Масса тела
            hue="species",
            kind="strip")
```

<div style="text-indent:30px; text-align:justify">Функция  <strong>catplot()</strong> с параметром  <strong>kind="strip"</strong> создает такую же диаграмму, как и функция  <strong>stripplot()</strong>. Недостатком такого графика является то, что точки на графике рассеяния перекрываются (сливаются).</div>

```python
sns.catplot(data=df, 
            x="species",     # Вид
            y="body_mass_g", # Масса тела
            hue="species",
            kind="swarm")
```

<div style="text-indent:30px; text-align:justify">Функция  <strong>catplot()</strong> с параметром  <strong>kind="swarm"</strong> создает такую же диаграмму, как и функция <strong>swarmplot()</strong>.</div>

<h4 style="text-align:center">Ящик с усами (коробчатый график)</h4>

<div style="text-indent:30px; text-align:justify">График разброса по категориям удобен для небольших наборов данных, так как по мере увеличения количества точек они начнут перекрывать друг друга и сливаться. </div> 
<div style="text-indent:30px; text-align:justify">Что бы преодолеть эти трудности, лучше воспользоваться графиками, которые сами содержат некоторую информацию о распределении внутри категорий. Один из таких графиков - это “<strong>ящик с усами</strong>”. Его можно построить с помощью той же функции  <strong>catplot()</strong>  с параметром  <strong>kind="box"</strong>. </div>

```python
sns.catplot(data=df, 
            x="species",     # Вид
            y="body_mass_g", # Масса тела
            hue="species",
            kind="box")
```

<div style="text-indent:30px; text-align:justify"><a href="https://ru.wikipedia.org/wiki/%D0%AF%D1%89%D0%B8%D0%BA_%D1%81_%D1%83%D1%81%D0%B0%D0%BC%D0%B8"><strong>График “ящик с усами”, или диаграмма размаха</strong></a>, был разработан Джоном Тьюки в 1970-х годах. По сути, ящик с усами – это быстрый способ изучения одного или нескольких наборов данных в графическом виде. Этот график может показаться более примитивным, чем, например, гистограммы, но он имеет некоторые преимущества. Он занимает меньше места и поэтому особенно полезен для сравнения распределений между несколькими группами или наборами данных. Кроме того, ящик с усами в своей первоначальной форме прост для построения.</div>

<h4 style="text-align:center">Скрипичная диаграмма</h4>

<div style="text-indent:30px; text-align:justify">Скрипичные диаграммы используются для визуализации распределения данных, отображая диапазон данных, медиану и область распределения данных.</div>
<div style="text-indent:30px; text-align:justify">Такие диаграммы, как и ящики с усами, показывают сводную статистику. Дополнительно они включают в себя графики плотности распределения, которые и определяют форму/распределение данных при визуализации.</div>
<div style="text-indent:30px; text-align:justify">Скрипичную диаграмму можно построить с помощью той же функции <strong>catplot()</strong> с параметром  <strong>kind="violin"</strong>:</div>

```python
sns.catplot(data=df, 
            x="species",     # Вид
            y="body_mass_g", # Масса тела
            hue="species",
            kind="violin")
```

<h4 style="text-align:center">Расширенная коробчатая диаграмма</h4>

<div style="text-indent:30px; text-align:justify">Боксеновская диаграмма (<strong>Boxen Plot</strong>), также известная как диаграмма буквенного значения (<strong>Letter-value Plot</strong>), представляет собой усовершенствованную коробчатую диаграмму, предназначенную для больших наборов данных (больше 10000). Внешне она похожа на традиционную коробчатую диаграмму. По сути, она просто отображает больше <a href="https://ru.wikipedia.org/wiki/%D0%9A%D0%B2%D0%B0%D0%BD%D1%82%D0%B8%D0%BB%D1%8C"><strong>квантилей</strong></a>. С большим количеством квантилей мы можем видеть больше информации о форме распределения за пределами центральных 50% данных, эта дополнительная деталь особенно присутствует в «хвостах», где коробчатые диаграммы, как правило, дают ограниченную информацию.</div>

```python
sns.catplot(data=df, 
            x="species",     # Вид
            y="body_mass_g", # Масса тела
            hue="species",
            kind="boxen")
```

<h4 style="text-align:center">Столбчатая диаграмма по значению</h4>

<div style="text-indent:30px; text-align:justify">Функция <strong>catplot()</strong> с параметром <strong>kind="bar"</strong> создает такую же столбчатую диаграмму, как и функция <strong>barplot()</strong>.</div>
<div style="text-indent:30px; text-align:justify">Высота бара (столбца) соответствует численному значению оценки категории (математическое ожидание, медиана и т.п.), линия, пересекающая верхнюю границу бара – показывает доверительный интервал:</div>

```python
sns.catplot(data=df, 
            x="species",     # Вид
            y="body_mass_g", # Масса тела
            hue="species",
            kind="bar")
```

<h4 style="text-align:center">Столбчатая диаграмма по количеству</h4>

<div style="text-indent:30px; text-align:justify">Функция <strong>catplot()</strong> с параметром <strong>kind="count"</strong> создает такую же столбчатую диаграмму,  как и функция <strong>countplot()</strong>. Для этой функции задается только один категориальный параметр <strong>x</strong>.</div>
<div style="text-indent:30px; text-align:justify">Функция определяет количество значений в каждой категории, и отображает полученное значение в виде столбчатой диаграммы:</div>

```python
sns.catplot(data=df, 
            x="species",     # Вид
            hue="species",
            kind="count")
```

<div style="text-indent:30px; text-align:justify">Рассмотренные выше три функции для построения <i>реляционных графиков</i> <strong>relplot()</strong>, графиков  <i>статистических распределений</i> <strong>displot()</strong>  и   <i>категориальных</i>  графиков <strong>catplot()</strong> можно свести в следующий рисунок:</div>

<div style="text-align:center"><img src="sb-func.png"></div>

<div style="text-indent:30px; text-align:justify">Эти три функции позволяют отобразить разнообразные виды диаграмм и заменяют более десятка отдельных функций (которые приведены в нижней части рисунка).</div>

<h4 style="text-align:center">Композиция из нескольких видов графиков (гибридный график)</h4>

<div style="text-indent:30px; text-align:justify">Крайне полезной является визуализация совместного распределения двух величин и визуализация распределений каждой из них в отдельности. Для этого можно воспользоваться функцией <strong>jointplot()</strong>, которая как раз и объединяет несколько видов графиков:</div>

```python
sns.jointplot(data=df, 
            x="body_mass_g",       # Масса тела
            y="flipper_length_mm", # Длина крыльев (плавников)
            hue="sex",
            kind="scatter")
```

<h4 style="text-align:center">Визуализация регрессионных моделей</h4>

<div style="text-indent:30px; text-align:justify">Для визуализации регрессионных моделей в Seaborn есть две функции: <strong>regplot()</strong> и <strong>lmplot()</strong>. Первая функция  строит график на уровне области <strong>Axes</strong> <i>Matplotlib</i> и принимает любые последовательности данных (списки, кортежи, массивы <i>NumPy</i> и одномерные объекты <i>Series Pandas</i>). Более абстрактная функция <strong>lmplot()</strong> принимает данные только в виде объектов <i>DataFrame Pandas</i> и строит график на уровне области <strong>Figure</strong> <i>Matplotlib</i>: </div>

```python
sns.lmplot(data=df, 
            x="flipper_length_mm", # Длина крыльев (плавников)
            y="body_mass_g")       # Масса тела
```

<div style="text-indent:30px; text-align:justify">Если воспользоваться функцией <strong>jointplot()</strong> с параметром <strong>kind="reg"</strong>, то можно изобразить регрессию с дополнительной информацией для каждой переменной: </div>

```python
sns.jointplot(data=df, 
            x="flipper_length_mm", # Длина крыльев (плавников)
            y="body_mass_g",       # Масса тела
            kind="reg")
```

<h4 style="text-align:center">Построение 3D-диаграммы рассеивания</h4>

<div style="text-indent:30px; text-align:justify">К сожалению, в <i>Seaborn</i> отсутствует 3D-движок. Являясь лишь дополнением к <i>Matplotlib</i>, он опирается на графические возможности основной библиотеки. Поэтому, для построения трехмерной диаграммы придется использовать уже знакомую вам функцию  <strong>scatter()</strong> из <i>Matplotlib</i>. Параметрам функции <strong>x</strong>, <strong>y</strong> и <strong>z</strong> присвоим значения соответствующих столбцов датафрейма <strong>df</strong>.</div>
<div style="text-indent:30px; text-align:justify">Вот как будет выглядеть точечный 3D-график с выборкой по уровням счастья, экономики и здоровья:</div>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение датасета
df = pd.read_csv('WHR20_DataForFigure2.1.csv')

# Русские названия некоторых категорий
df = df.rename(columns={'Logged GDP per capita': 'ВВП',
                        'Healthy life expectancy': 'Продолжительность жизни',
                        'Regional indicator': 'РЕГИОН',
                        'Social support': 'Социальная поддержка',
                        'Ladder score':'Оценка счастья'})

fig = plt.figure()                      # Создать рисунок
ax = fig.add_subplot(projection = '3d') # Добавить 3D область

x = df['Оценка счастья']
y = df['ВВП']
z = df['Продолжительность жизни']

ax.set_xlabel("Счастье")
ax.set_ylabel("Экономика")
ax.set_zlabel("Здоровье")

ax.scatter(x, y, z, c='red', s=100)     # Точечный график
```
