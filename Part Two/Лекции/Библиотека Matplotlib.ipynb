<h1 style="text-align:center">Библиотека Matplotlib</h1>
<h2 style="text-align:center">Что такое Matplotlib и зачем она нужна</h2>
<div style="text-indent:30px; text-align:justify">Matplotlib является основным пакетом для визуализации данных в Python. Он позволяет работать с данными на нескольких уровнях:</div>
<div style=" margin-left:50px">
<li style="text-align:justify">с помощью модуля <strong>Pyplot</strong>, который рассматривает график как единое целое; </li> 
<li style="text-align:justify">через объектно-ориентированный интерфейс, когда каждая фигура или ее часть является отдельным объектом, — это позволяет выборочно менять их свойства и отображение.</li>
</div>

<div style="text-indent:30px; text-align:justify">Matplotlib используют для визуализации данных любой сложности. Библиотека Matplotlib позволяет строить разные варианты графиков: линейные, трехмерные, диаграммы рассеяния и другие, а также комбинировать их. </div> 
<div style="text-indent:30px; text-align:justify">Сама Matplotlib является основой для других библиотек — например, Seaborn позволяет проще создавать графики и имеет больше возможностей для косметического улучшения их внешнего вида.  </div>

<h2 style="text-align:center">Как связаны Matplotlib, Pyplot и Pylab</h2>

<div style="text-indent:30px; text-align:justify">При погружении в Matplotlib можно встретить упоминание двух модулей — Pyplot и Pylab. Важно понимать, какой из них использовать в работе и почему они появились. Разберемся в терминологии.</div>

<div style="text-indent:30px; text-align:justify">Библиотека Matplotlib — это пакет для визуализации данных в Python. <strong>Pyplot</strong> — это модуль в пакете Matplotlib. Его вы часто будете видеть в коде как matplotlib.pyplot. Модуль помогает автоматически создавать оси, фигуры и другие компоненты, не задумываясь о том, как это происходит. Именно Pyplot используется в большинстве случаев.</div>

<div style="text-indent:30px; text-align:justify">Pylab — это еще один модуль, который устанавливается вместе с пакетом Matplotlib. Он одновременно импортирует Pyplot и библиотеку NumPy для работы с массивами в интерактивном режиме или для доступа к функциям черчения при работе с данными.</div>

<div style="text-indent:30px; text-align:justify">Сейчас Pylab имеет только историческое значение — он облегчал переход с MATLAB на Matplotlib, так как позволял обходиться без операторов импорта (а именно так привыкли работать пользователи MATLAB). Вы можете встретиться с Pylab в примерах кода на разных сайтах, но на практике использовать модуль не придется.</div>

<h2 style="text-align:center">Название, оси, сетка графика, тип и цвет линии</h2>
<div style="text-indent:30px; text-align:justify">Теперь дадим графику название, подпишем оси и отобразим сетку. </div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 30)      # Создание массива значений аргумента X

# Построение графика функции
plt.plot(x, np.sin(x))         # Построение графика
plt.title('Выходной сигнал')   # Название графика
plt.grid()                     # Показать сетку
plt.xlabel('Время, сек')       # Наименование оси абсцисс
plt.ylabel('Амплитуда, Вольт') # Наименование оси ординат
```

<div style="text-indent:30px; text-align:justify">Изменим тип линии и ее цвет, для этого в функцию <strong>plot()</strong>, в качестве третьего параметра, передадим строку, сформированную определенным образом, в нашем случае это <strong>'r--'</strong>, где <strong>r</strong> означает красный цвет, а <font size="+2">--</font> определяет тип линии – пунктирная линия. </div>
<div style="text-indent:30px; text-align:justify">Подробнее об использовании символов, задающих стиль линии и цвет, а также вид маркера, можно  прочитать <a href="https://jenyay.net/Matplotlib/LineStyle"><strong>здесь</strong></a>.</div> 
<div style="text-indent:30px; text-align:justify">Функция  <strong>sin(x)</strong> из модуля  numpy  создает новый массив, каждый элемент которого является синусом соответствующего элемента исходного массива <strong>x</strong>.</div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 30)      # Создание массива значений аргумента X

# Построение графика функции
plt.plot(x, np.sin(x), 'r--')  # Построение графика
plt.title('Выходной сигнал')   # Название графика
plt.grid()                     # Показать сетку
plt.xlabel('Время, сек')       # Наименование оси абсцисс
plt.ylabel('Амплитуда, Вольт') # Наименование оси ординат
```

<h2 style="text-align:center">Несколько графиков на одном поле</h2>
<div style="text-indent:30px; text-align:justify">Построим несколько графиков на одном поле, для этого добавим квадратичную зависимость.</div>
<div style="text-indent:30px; text-align:justify">Функция  <strong>square(x)</strong> из модуля  numpy  создает новый массив, каждый элемент которого является квадратом числа соответствующего элемента исходного массива&nbsp;<strong>x</strong>. 

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100500)      # Создание массива значений аргумента X

# Построение графиков двух функций на одном рисунке
plt.plot(x, np.sin(x), x, np.square(np.sin(x)), 'r--')
plt.title('Выходные сигналы')  # Название графика
plt.grid()                     # Показать сетку
plt.xlabel('Время, сек')       # Наименование оси абсцисс
plt.ylabel('Амплитуда, Вольт') # Наименование оси ординат
```

<div style="text-indent:30px; text-align:justify">В приведенном примере в функцию <strong>plot()</strong> последовательно передаются два массива для построения первого графика и два массива для построения второго графика, при этом, для обоих графиков массив значений независимой переменной <strong>x</strong> один и то же.</div>

<h2 style="text-align:center">Представление графиков на разных полях</h2>
<div style="text-indent:30px; text-align:justify">Третья, довольно часто встречающаяся задача – это отобразить два или более различных поля, на которых будет отображено по одному или более графику. Построим уже известные нам две зависимость на разных полях:</div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 30)         # Создание массива значений аргумента X

# Построение графиков функций
# DPI=157 плотность точек для 14" экрана Full HD (1920×1080)
# DPI=92 плотность точек для 32" экрана QHD (2560×1440)
plt.figure(figsize=(8,10), dpi=92, facecolor="LightGray") # Ширина и высота рисунка в дюймах, цвет фона светлосерый

# Построение графика 1 
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x)) 
plt.title('Графики функций')      # Заголовок
plt.xlabel('x', fontsize=14)      # Ось абсцисс
plt.ylabel('f1(x)', fontsize=14)  # Ось ординат
plt.grid(True)                    # Показать сетку

# Построение графика 2
plt.subplot(2, 1, 2)
plt.plot(x, np.sin(x)-np.square(np.sin(x)))
plt.xlabel('x', fontsize=14)      # Ось абсцисс
plt.ylabel('f2(x)', fontsize=14)  # Ось ординат
plt.grid(False)                   # Показать сетку
```

<div style="text-indent:30px; text-align:justify">Здесь мы воспользовались новыми функциями:</div>
<ul>
<li style="margin-left:20px; text-align:justify"><strong>figure()</strong> – функция для задания глобальных параметров отображения графиков. Вот ее некоторые параметры:</li>
<ul type="circle">
<li style="margin-left:20px; text-align:justify"><strong>figsize</strong> – указываем кортеж размера фигуры в дюймах, по умолчанию равно (6.4, 4.8), т.е. 640х480 пикселей;</li>
<li style="margin-left:20px; text-align:justify"><strong>dpi</strong> – количество точек на дюйм (плотность пикселей), по умолчанию равно 100.0. <br>
Для мониторов, экранов смартфонов,  телевизоров и других цифровых экранов правильнее говорить о характеристике PPI (количество пикселей на дюйм).  DPI же применяется для указания плотности точек при выводе информации на плоский носитель, например, при печати принтером на бумаге или же при сканировании. Вы можете самостоятельно рассчитать PPI для своего монитора, воспользовавшись онлайн калькулятором;</li>
<li style="margin-left:20px; text-align:justify"><strong>facecolor</strong> - цвет фона фигуры, по умолчанию задано white.</li>
</ul>
<li style="margin-left:20px; text-align:justify"><strong>subplot()</strong> – функция для задания местоположения поля с графиком.</li>
</ul>    

<div style="text-indent:30px; text-align:justify">Существует несколько способов задания областей для вывода графиков. В примере мы воспользовались вариантом, который предполагает передачу трех аргументов: первый аргумент - количество строк, второй - столбцов в формируемом поле, третий - индекс (номер поля, считаем сверху вниз, слева направо). </div>

<div style="text-indent:30px; text-align:justify">Остальные функции вам знакомы, дополнительно мы использовали параметр <strong>fontsize</strong> функций <strong>xlabel()</strong> и <strong>ylabel()</strong> для задания размера шрифта.</div>
