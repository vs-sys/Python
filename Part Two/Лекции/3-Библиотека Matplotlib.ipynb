<h2 style="text-align:center">Столбчатые диаграммы</h2>

<div style="text-indent:30px; text-align:justify">Столбчатые диаграммы наиболее эффективны тогда, когда вам необходимо визуализировать данные в виде категорий, если их число не превышает 10. Если у нас слишком много категорий, то столбцы будут сильно загромождать график. </div>
<div style="text-indent:30px; text-align:justify; margin-top:10px">Для их построения используются две функции:</div>

<div style=" margin-left:50px">
<li style="text-align:justify"><strong>bar()</strong> – вертикальная столбчатая диаграмма;</li> 
<li style="text-align:justify"><strong>barh()</strong> – горизонтальная столбчатая диаграмма.</li>
</div>

<div style="text-indent:30px; text-align:justify; margin-top:10px">Построим простую столбчатую диаграмму:</div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

# Набор данных
np.random.seed(23) # Начальные условия для генератора случайных чисел
groups = [f'P{i}' for i in range(7)] # Список категорий через F-строку
counts = np.random.randint(3, 10, len(groups)) # Массив значений
plt.bar(groups, counts)              # Вертикальная столбчатая диаграмма

plt.show() # Отображение диаграммы
```

<div style="text-indent:30px; text-align:justify">Если заменим  <strong>bar()</strong> на  <strong>barh()</strong>, то получим горизонтальную диаграмму:</div>
<div style="text-indent:60px; text-align:justify"><strong>plt.barh(groups, counts)</strong></div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

# Набор данных
np.random.seed(23) # Начальные условия для генератора случайных чисел
groups = [f'P{i}' for i in range(7)] # Список категорий через F-строку
counts = np.random.randint(3, 10, len(groups)) # Массив значений
plt.barh(groups, counts)              # Вертикальная столбчатая диаграмма

plt.show() # Отображение диаграммы
```

<div style="text-indent:30px; text-align:justify"><u>Пояснения.</u></div>
<div style="text-indent:30px; text-align:justify">Функция  <strong>random.seed()</strong> из модуля  <strong>numpy</strong>  задает начальные условия для генератора случайных чисел. Генератор, инициализированный одними и теми же начальными условиями, выдает абсолютно одинаковые случайные последовательности. Параметром является целое положительное число в интервале от <strong>0</strong> до  <strong>2<sup>32</sup></strong>.</div> 
<div style="text-indent:30px; text-align:justify">В Python, начиная с версии 3.6, появился новый тип строк – f-строки, которые буквально означают “formatted string”. F-строки задаются с помощью литерала «f» перед кавычками. F-строки делают очень простую вещь – они берут значения переменных, которые есть в текущей области видимости, и подставляют их в строку. В самой строке вам лишь нужно указать имя этой переменной в фигурных скобках.</div>

<div style="text-indent:30px; text-align:justify; margin-top:10px">Например:</div>

```python
name = "дмитрий иванов"
age = 25
# Метод title, делает первую букву каждого слова заглавной
print(f"Меня зовут {name.title()} и мне {age} лет") 
```

<div style="text-indent:30px; text-align:justify">Функция  <strong>random.randint()</strong> из модуля  <strong>numpy</strong>  задает массив случайных целых чисел из заданного интервала. Если третий параметр не указан, то будет возвращено одно целое случайное число. Если в качестве третьего параметра указано одно число, то будет возвращен одномерный массив с заданным количеством элементов. Если указано несколько чисел, то количество этих чисел будет определять количество осей массива, а их значения - количество элементов вдоль каждой из них.</div>

<h3 style="text-align:center">Дополнительные параметры функций <strong>bar()</strong> и  <strong>barh()</strong></h3>

<div style=" margin-left:50px">
<li style="text-align:justify"><strong>color</strong>  – цвет(а) столбцов диаграммы;</li> 
<li style="text-align:justify"><strong>edgecolor</strong> – цвет(а) границ столбцов;</li>
<li style="text-align:justify"><strong>linewidth</strong> – ширина границы (скалярная величина или массив);</li>
<li style="text-align:justify"><strong>tick_label</strong> – метки для столбов (строка или массив строк);</li>
<li style="text-align:justify"><strong>xerr, yerr</strong> – величина ошибки для графика (скалярная величина или массив);</li>
<li style="text-align:justify"><strong>ecolor</strong> – цвет линий ошибки</li>
<li style="text-align:justify"><strong>log</strong> – логарифмический масштаб для оси y (по умолчанию False);</li>
</div>

<div style="text-indent:30px; text-align:justify; margin-top:10px">Пример, демонстрирующий работу с параметрами  <strong>bar()</strong>:</div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)  # Начальные условия для генератора случайных чисел
groups = [f"P{i}" for i in range(7)]  # Список категорий через F-строку
counts = np.random.randint(0, 10, len(groups))       # Массив значений
bw = counts * 0.1                      # Ширина пропорциональна высоте
bc=["b","g","r","c","m","y","k"]       # Список цветов
plt.bar(groups, counts, width=bw, color=bc, alpha=0.6, edgecolor='blue', linewidth=5)

plt.show() # Отображение диаграммы
```

<h2 style="text-align:center">Групповые столбчатые диаграммы</h2>
<div style="text-indent:30px; text-align:justify">Используя определенным образом подготовленные данные, можно строить групповые диаграммы:</div>


```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

categories = [f'P{i}' for i in range(6)] # Список категорий через F-строку

x = np.arange(len(categories))           # Номера категорий

g1 = [10, 21, 34, 12, 27, 9]   # Первый массив значений
g2 = [17, 15, 25, 21, 26, 13]  # Второй массив значений

width = 0.3                    # Ширина прямоугольников

plt.bar(categories, g1, width, label='G1') # Первая диаграмма с категориями
plt.bar(x+width, g2, width, label='G2')    # Вторая сдвинутая диаграмма
plt.legend()

plt.show() # Отображение диаграммы
```

<h2 style="text-align:center">Гистограммы</h2>
<div style="text-indent:30px; text-align:justify">Гистограмма состоит из примыкающих прямоугольников, расположенных вдоль оси <strong>x</strong>, которые разбиты на дискретные интервалы, которые называют бинами (<strong>bins</strong>). Их площадь пропорциональна частоте конкретного интервала. Общая площадь гистограммы равна количеству данных. Такой способ визуализации часто используют в статистике для демонстрации распределения.</div>
<div style="text-indent:30px; text-align:justify">Для представления гистограммы используется функция <strong>hist()</strong>. Она не только рисует гистограмму, но также может возвратить кортеж значений, представляющих собой результат вычислений гистограммы. </div>
<div style="text-indent:30px; text-align:justify">В качестве примера, сгенерируем набор из 10000 случайных чисел из <a href="https://ru.wikipedia.org/wiki/%D0%9D%D0%BE%D1%80%D0%BC%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5"><strong>нормального распределения (Гаусса)</strong></a> при среднем значении <strong>μ = 0</strong> и стандартной девиации  <strong>σ = 1</strong>  с помощью функции <strong>np.random.normal()</strong> из модуля <strong>numpy</strong>. Дальше создадим гистограмму из этих данных, передавая аргумент функции <strong>hist()</strong>. Например, нужно разделить данные на 30 интервалов (значение по умолчанию – 10 интервалов). Для этого используется именованный аргумент <strong>bins</strong>:</div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0, 1, size=10000) # Набор чисел

plt.hist(data, bins=30) # Гистограмма с 30 бинами

plt.show() # Отображение гистограммы
```

<h2 style="text-align:center">Классические круговые диаграммы</h2>
<div style="text-indent:30px; text-align:justify">Круговые диаграммы – это наглядный способ показать доли  компонентов в наборе. Они идеально подходят для отчетов, презентаций и публикаций. Для построения круговых диаграмм в <strong>Matplotlib</strong> используется функция <strong>pie()</strong>.</div>

```python
# Импорт библиотеки
import matplotlib.pyplot as plt

fruits = ['Яблоки', 'Киви', 'Бананы', 'Лимоны', 'Арбузы']  # Список меток
counts = [34, 25, 43, 31, 17]                              # Значения долей

# Построение круговой диаграммы
plt.pie(counts, labels=fruits)

plt.title('Количество фруктов')

plt.show() # Отображение диаграммы
```

<div style="text-indent:30px; text-align:justify">Ниже перечислены только некоторые из всех параметров функции  <strong>pie()</strong>:</div>

<div style=" margin-left:50px">
<li style="text-align:justify">Массив с размерами долей (обязательный параметр);</li> 
<li style="text-align:justify"><strong>labels</strong> – список текстовых меток долей (по умолчанию None);</li>
<li style="text-align:justify"><strong>colors</strong> – массив цветов долей (по умолчанию None);</li>
<li style="text-align:justify"><strong>explode</strong> – если параметр не равен None (это значение по умолчанию), то часть долей, которые перечислены в передаваемом значении, будут вынесены из диаграммы на заданное расстояние;</li>
<li style="text-align:justify"><strong>autopct</strong> – формат текстовой метки внутри доли (по умолчанию None);</li>
<li style="text-align:justify"><strong>shadow</strong> – отображение тени для диаграммы (по умолчанию False);</li>
<li style="text-align:justify"><strong>wedgeprops</strong> – словарь параметров, определяющих внешний вид долей (по умолчанию None).</li>
</div>

<div style="text-indent:30px; text-align:justify; margin-top:10px">Демонстрация работы некоторых параметров функции  <strong>pie()</strong>:</div>

```python
# Импорт библиотеки
import matplotlib.pyplot as plt

fruits = ['Яблоки', 'Киви', 'Бананы', 'Лимоны', 'Арбузы']  # Список меток
counts = [34, 25, 43, 31, 17]                              # Значения долей

explode = [0.2, 0, 0.2, 0, 0]          # Значения вынесения различных долей

# Построение круговой диаграммы
plt.pie(counts, labels=fruits, explode=explode, autopct='%1.2f%%', shadow=True, 
         wedgeprops={'lw':2, 'ls':'--','edgecolor':'k'}, rotatelabels=True)

plt.title('Количество фруктов', loc='left')

plt.show() # Отображение диаграммы
```

<h2 style="text-align:center">Вложенные круговые диаграммы</h2>
<div style="text-indent:30px; text-align:justify">Вложенная круговая диаграмма состоит из двух компонентов. Внутренняя часть является детальным представлением информации, а внешняя – суммарной по заданным областям. Каждая область представляет собой список численных значений, вместе они образуют общий набор данных.</div>
<div style="text-indent:30px; text-align:justify; margin-top:10px">Пример:</div>

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

# Набор данных
data = np.array([[5, 10, 7], [8, 15, 5], [11, 9, 7]])

# Наборы цветов
colors_1 = ['r', 'g', 'b']
colors_2 = ['c', 'y', 'm']

# Построение внешней круговой диаграммы
# Метод sum(axis=1) вычисляет суммы в строках
plt.pie(data.sum(axis=1), radius=1, colors=colors_1, wedgeprops={'width':0.5, 'edgecolor':'w'})

# Построение внутренней круговой диаграммы
# Метод flatten() возвращает копию массива сжатую до одного измерения
plt.pie(data.flatten(), radius=0.6, colors=colors_2, wedgeprops=dict(width=0.4, edgecolor='k'))

plt.show() # Отображение диаграмм
```

<h2 style="text-align:center">Кольцевая  диаграмма</h2>
<div style="text-indent:30px; text-align:justify; margin-top:10px">Для того чтобы построить круговую диаграмму с отверстием, необходимо параметру <strong>wedgeprops</strong>, отвечающему за внешний вид долей, задать значение <strong>dict(width=0.5)</strong>:</div>

```python
# Импорт библиотеки
import matplotlib.pyplot as plt

# Наборы данных
sales = [24, 17, 53, 21, 9]
models = ['Ford', 'Toyota', 'BMV', 'AUDI', 'Mercedes']

# Построение кольцевой диаграммы
plt.pie(sales, labels=models, wedgeprops=dict(width=0.5, edgecolor='w'))

plt.show() # Отображение диаграммы
```

<h2 style="text-align:center">Графики в полярной системе координат</h2>

<div style="text-indent:30px; text-align:justify">По умолчанию двумерные графики имеют прямоугольную декартову систему координат. В некоторых задачах намного нагляднее отображать информацию на круге, а не в прямоугольнике. Например, при отображении розы ветров. Переход от прямоугольной формы основы рисунка к кругу позволяет нагляднее отобразить графики, заданные в полярной системе координат.</div>

<div style="text-indent:30px; text-align:justify">В декартовой системе координат (используется по умолчанию), положение точки определяется двумя значениями: аргументом <strong>x</strong> и значением функции <strong>y</strong>.</div>

<div style="text-indent:30px; text-align:justify">В полярной системе координат положение точки также определяется двумя значениями: радиусом (расстояние от начала координат) <strong>$r$</strong> и углом <strong>$\varphi$</strong>. Угол обычно отсчитывают против часовой стрелке от "восточного" (3:00) направления.</div>

<div style="text-indent:30px; text-align:justify">Переход из полярной системы координат в прямоугольную выглядит так:</div>

<div style="text-indent:60px; text-align:justify"><strong>x = r*cos($\varphi$)</strong></div>
<div style="text-indent:60px; text-align:justify"><strong>y = r*sin($\varphi$)</strong></div>

<div style="text-indent:30px; text-align:justify">Для построения кривой в полярных координатах используется функция <strong>polar()</strong>.</div>
<div style="text-indent:30px; text-align:justify; margin-top:10px">Вот пример построения спирали (первый массив — <strong>$\varphi$</strong>, второй — <strong>$r$</strong>):</div>

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

# Набор данных
t = np.linspace(0, 4 * np.pi, 100)

# График в полярной системе координат
plt.polar(t, t, linewidth=6)

plt.show()
```

<div style="text-indent:30px; text-align:justify">А это угловое распределение пионов в <strong>$e^+ e^-$</strong> аннигиляции:</div>

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

# Набор данных
phi = np.linspace(0, 2 * np.pi, 100)

# График в полярной системе координат
plt.polar(phi, np.sin(phi) ** 2, linewidth=6)

plt.show()
```

<h2 style="text-align:center">Контурные графики</h2>
<div style="text-indent:30px; text-align:justify">Пусть мы хотим изучить поверхность z=xy.</div>
<div style="text-indent:30px; text-align:justify; margin-top:10px">Вот горизонтали этой поверхности:</div>

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

# Наборы данных
x = np.linspace(-1, 1, 50)
y = x
z = np.outer(x, y)  # Функция из модуля numpy, вычисляет внешнее произведене двух заданных векторов 

plt.contour(x, y, z) # Функция рисует линии уровня

plt.show()
```

<div style="text-indent:30px; text-align:justify">Что-то их маловато. Сделаем побольше и подпишем:</div>

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

curves = plt.contour(x, y, z, np.linspace(-1, 1, 30))
plt.clabel(curves)                 # Функция добавляет подписи к линиям уровня
plt.title(r'$z=xy$', fontsize=20)  # Заголовок рисунка

plt.show()
```

<div style="text-indent:30px; text-align:justify">А здесь высота дается цветом, как на физических географических картах. Функция <strong>colorbar</strong> показывает соответствие цветов и значений z.</div>

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

plt.contourf(x, y, z, np.linspace(-1, 1, 40))  # Функция закрашивает области между линиями уровня
plt.colorbar()  # Функция показывает соответствие цветов и значений z

plt.show()
```
