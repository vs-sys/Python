#### __О наборе данных__
Загрязнение воздуха – это загрязнение внутренней или внешней среды любым химическим, физическим или биологическим агентом, который изменяет естественные характеристики атмосферы. Бытовые устройства сжигания, автомобили, промышленные объекты и лесные пожары являются распространенными источниками загрязнения воздуха. Загрязнители, представляющие серьезную опасность для здоровья населения, включают твердые частицы, окись углерода, озон, двуокись азота и двуокись серы. Загрязнение атмосферного воздуха и воздуха внутри помещений вызывает респираторные и другие заболевания и является важным источником заболеваемости и смертности.

##### __Данные содержат следующие характеристики:__
* __Country__: Название страны
* __City__: Название города
* __AQI Value__: Индекс качества воздуха (AQI) используется для информирования населения о том, насколько загрязнен воздух в настоящее время или насколько загрязненным он по прогнозам станет. Индекс качества воздуха колеблется от 0 до 500, хотя качество воздуха может индексироваться выше 500 при более высоких уровнях опасного загрязнения воздуха. Хорошее качество воздуха колеблется от 0 до 50, а измерения выше 300 считаются опасными.
* __AQI Category__: Категории AQI по загрязнению
* __CO AQI Value__: Монооксид углерода (Угарный газ) - это бесцветный газ без запаха. На улице он выделяется в воздух, прежде всего, легковыми и грузовыми автомобилями, а также другими транспортными средствами или механизмами, сжигающими ископаемое топливо. Такие предметы, как керосиновые и газовые обогреватели, газовые плиты, также выделяют CO, влияющий на качество воздуха в помещении. Вдыхание воздуха с высокой концентрацией CO снижает количество кислорода, который может транспортироваться в кровотоке к таким важным органам, как сердце и мозг. При очень высоких уровнях, которые маловероятны на открытом воздухе, но возможны в закрытых помещениях. CO может вызвать головокружение, спутанность сознания, потерю сознания и смерть
* __CO AQI Category__: Категория AQI угарного газа в городе
* __Ozone AQI Value__: Озон [O3]. Молекула озона вредна для качества наружного воздуха (если он находится за пределами озонового слоя). На уровне поверхности озон образуется в результате химических реакций между оксидами азота и летучими органическими соединениями (ЛОС). В отличие от хорошего озона, находящегося в верхних слоях атмосферы, озон на уровне земли может спровоцировать ряд проблем со здоровьем, таких как боль в груди, кашель, раздражение горла и воспаление дыхательных путей. Кроме того, он может снизить функцию легких и ухудшить течение бронхита, эмфиземы и астмы. Озон также влияет на растительность и экосистемы. В частности, он повреждает чувствительную растительность в период вегетации
* __Ozone AQI Category__: Категории озона по загрязнению
* __NO2 AQI Value__: Диоксид азота [NO2]. Диоксид азота является одним из нескольких оксидов азота. Он попадает в воздух в результате природных явлений, таких как попадание из стратосферы или освещение. Однако на уровне поверхности NO2 образуется в результате выхлопов легковых автомобилей, грузовиков и автобусов, электростанций и внедорожной техники. Воздействие в течение коротких периодов времени может привести к обострению респираторных заболеваний, таких как астма. Более длительное воздействие может способствовать развитию астмы и респираторных инфекций. Люди с астмой, дети и пожилые люди подвержены большему риску воздействия NO2 на здоровье
* __NO2 AQI Category__: Категории диоксида азота по загрязнению
* __PM2.5 AQI Value__: Твердые частицы [PM2.5]. Атмосферные твердые частицы, также известные как атмосферные аэрозольные частицы, представляют собой сложные смеси мелких твердых и жидких частиц, попадающих в воздух. При вдыхании они могут вызвать серьезные проблемы с сердцем и легкими. Международным агентством по изучению рака (IARC) они были классифицированы как канцероген группы 1. PM10 относится к частицам диаметром 10 микрометров и менее. PM2.5 относится к частицам диаметром 2,5 микрометра или меньше.
* __PM2.5 AQI Category__: Категории твердых частиц по загрязнению

_**Для удобства преобразуем их в более короткие названия и аббревиатуры**_

```python
for pair in [
    ["Russian Federation", "Russia"],
    ["United States of America", "USA"],
    ["Republic of North Macedonia", "North Macedonia"],
    ["United Kingdom of Great Britain and Northern Ireland", "UK"],
    ["United Republic of Tanzania", "Tanzania"],
    ["Iran (Islamic Republic of)", "Iran"],
    ["Democratic Republic of the Congo", "Congo"],
    ["Papua New Guinea", "PNG"],
    ["Bosnia and Herzegovina", "BiH"],
    ["Dominican Republic", "Dominicana"],
    ["Bolivia (Plurinational State of)", "Bolivia"],
    ["Venezuela (Bolivarian Republic of)", "Venezuela"],
    ["Republic of Moldova", "Moldova"],
    ["Kingdom of Eswatini", "Eswatini"],
    ["Central African Republic", "CAR"],
    ["Republic of Korea", "Korea"],
    ["United Arab Emirates", "UAE"],
    ["Syrian Arab Republic", "Syria"],
    ["Lao People's Democratic Republic", "Laos"],
    ["Trinidad and Tobago", "T&T"],
    ["Equatorial Guinea", "GQ"],
    ["State of Palestine", "Palestine"],
    ["Saint Kitts and Nevis", "SK&N"]]:

    df['Country'] = df['Country'].replace(pair[0], pair[1], regex=False) 
```

___По каким странам в наборе данных наибольшее количество наблюдений___

```python
# Выведем топ стран на столбиковой диаграмме
plt.figure(figsize=(8,8))
df.groupby('Country')['AQI Category'].count().sort_values(ascending=False).head(25).sort_values(ascending=True).plot(kind='barh', color='#5555FF')
plt.title('\nНаиболее представленные страны\n', fontsize=20)
plt.ylabel('Страна')
plt.xlabel('Количество наблюдений')
plt.show()
```

___Посмотрим категории и соответствующие им числовые значения___

```python
sns.catplot(data=df, x="AQI Category", y="AQI Value", kind="box", height=4, aspect=2.4)
plt.xlabel("Категория", fontsize=18);
plt.ylabel("Значение", fontsize=18);
```

___Построим график в котором отражается количество наблюдений для стран с наилучшим качеством воздуха___

```python
# Выберем строки данных, в которых AQI Category имеет значение Good 
top_low = df[df['AQI Category'] == 'Good']
top_good_list = top_low.groupby('Country')['AQI Category'].count().sort_values(ascending=False).head(15)

# Построим график 
top_good_list.plot(kind='bar', color='#5555FF', figsize=(8, 6)) 
```

По графику можно видеть, что категория __Good__ чаще встречается в __Бразилии__, __России__ и __США__

___Построим график в котором отражается количество наблюдений для стран с наихудшим качеством воздуха___

```python
# Выберем строки данных, в которых AQI Category имеет значение Hazardous 
top_high = df[df['AQI Category'] == 'Hazardous']
top_hazardous_list = top_high.groupby('Country')['AQI Category'].count().sort_values(ascending=False).head(15)

# Построим график 
top_hazardous_list.plot(kind='bar', color='#5555FF', figsize=(8, 6)) 
```

По графику можно видеть, что в __Индии__ наибольшее количество наблюдений с опасным качеством воздуха

___Выделим топ часто встречающихся стран и создадим диаграмму с накоплением по категориям AQI___

```python
top_list = (df.groupby('Country')
               ['AQI Category'].count()
               .sort_values(ascending=False)
               .head(15)
               .index
               .to_list())

# Сделаем подвыборку из набора данных с этими странами
data_top_country = df.loc[df['Country'].isin(top_list)]

# Построим диаграмму с накоплением
pd.crosstab(data_top_country['Country'], data_top_country['AQI Category']).plot.bar(stacked=True)
```
