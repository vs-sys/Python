<h2 style="text-align:center">Введение</h2>

<div style="text-indent:30px; text-align:justify"><strong>Seaborn</strong> – это эффективная библиотека для визуализации  данных  и выделения их статистических особенностей в Data Science. Она построена на основе библиотеки  <strong>matplotlib</strong>, но более высокоуровневая. Важная особенность библиотеки – это заложенный в нее механизм предобработки данных, возможный благодаря тесной интеграции с библиотекой <strong>pandas</strong>. Благодаря этому специалист по анализу данных (data scientist), который тратит 70-80% всего своего рабочего времени на исследование и подготовку данных, может увидеть корреляцию данных там, где она была не очевидна.</strong></div> 

<div style="text-indent:30px; text-align:justify">Две основные причины для использования:</div>
<div style=" margin-left:60px">
<li style="text-align:justify">Выразительный высокоуровневый интерфейс: построение большинства простых графиков происходит в одну строчку кода.</li>
<li style="text-align:justify">Более эстетичные графики: часто встроенные в seaborn стили достаточно хороши и без вашего вмешательства.</li>
</div>

<div style="text-indent:30px; text-align:justify">Данные можно передавать прямо в объектах <strong>DataFrame</strong> (это двухмерная структура, состоящая из столбцов и строк. У столбцов есть имена, а у строк – индексы). Библиотека сама выполняет всю необходимую работу: агрегацию (выделение подмножеств), статистическую обработку (например, вычисление доверительных интервалов) и визуальное выделение полученных результатов. </div>
<div style="text-indent:30px; text-align:justify">Seaborn можно использовать в следующих случаях:</div>
<div style=" margin-left:60px">
<li style="text-align:justify">вы знакомы с <strong>matplotlib</strong> и при этом хотите получить более красивые диаграммы;</li>
<li style="text-align:justify">ваш основной Data Science инструмент – это <strong>pandas</strong>, с которым seaborn интегрируется довольно просто;</li>
<li style="text-align:justify">нужна визуализация статистических графиков, например, линейной регрессии.</li>
</div>

<h2 style="text-align:center">Точечный график (Диаграмма рассеяния)</h2>
<div style="text-indent:30px; text-align:justify">Для загрузки датасета <strong>mpg</strong> будем использовать следующий код:

```python
cars = sns.load_dataset("mpg")
```

<div style="text-indent:30px; text-align:justify">Функция  <strong>load_dataset()</strong> скачивает из учебного репозитория набор данных и возвращает его в виде объекта <strong>cars</strong> (который является экземпляром класса <strong>DataFrame</strong>). В наборе <strong>mpg</strong> содержится информация о характеристиках ряда автомобилей (более подробную информацию об этом наборе данных вы можете прочитать <a href="https://www.kaggle.com/datasets/uciml/autompg-dataset"><strong>по этой ссылке</strong></a>). 
<div style="text-indent:30px; text-align:justify">Для того чтобы получить названия всех категорий (столбцов) в этом наборе данных, применим метод  <strong>columns</strong> к нашему датафрейму:</div>

```python
cars.columns
```

<div style="text-indent:30px; text-align:justify">Теперь, изучив названия категорий, построим, используя функцию <strong>relplot()</strong>, зависимость рабочего объема двигателя (<strong>displacement</strong>)  от  мощности двигателя в лошадиных силах (<strong>horsepower</strong>), при этом размер точки будет определяться  количеством цилиндров (<strong>cylinders</strong>):</div>

```python
sns.relplot(data=cars, x="horsepower", y="displacement", size="cylinders")
```

<div style="text-indent:30px; text-align:justify">Обратите внимание, что мы просто передали переменную <strong>cars</strong> параметру <strong>data</strong>, не выполняя никаких преобразований формата данных.</div>

> <h5 style="text-indent:30px; text-align:justify; background-color:LightGray"><u>Примечание.</u> Если вы до сих пор не используете продвинутую среду разработки (например, Jupyter Notebook, PyCharm, Spyder), то может потребоваться явный вызов функции  matplotlib.pyplot.show():</h5>

```python
import matplotlib.pyplot as plt
plt.show()
```

<div style="text-indent:30px; text-align:justify">Как видно, при помощи seaborn очень легко составлять простые графики наподобие диаграмм рассеивания. Нам не обязательно использовать объект Figure и экземпляры Axes или что-нибудь настраивать. Здесь мы передали <strong>DataFrame</strong> в качестве аргумента с данными, а признаки с информацией, которую нужно визуализировать, указали в параметрах  <strong>x</strong> и <strong>y</strong>.</div>
<div style="text-indent:30px; text-align:justify">Довольно интересной для поиска закономерностей в признаках данных является функция, которая строит графики парного сравнения параметров <strong>pairplot()</strong>. В эту функцию можно передать весь датафрейм целиком. </div>

```python
sns.pairplot(data=cars)
```

<div style="text-indent:30px; text-align:justify">Однако при большом количестве столбцов построение может занять весьма значительное время, а большое количество вариантов – затруднить восприятие:</div>
<div style="text-indent:30px; text-align:justify; margin-top:20px">Поэтому мы создадим новый датафрейм <strong>newcars</strong>, который содержит только три признака "<strong>mpg</strong>", "<strong>horsepower</strong>", "<strong>displacement</strong>":</div>

```python
newcars = cars[["mpg", "horsepower", "displacement"]]  # Новый датафрейм с тремя признаками (столбцами)
sns.pairplot(data=newcars)
```

<div style="text-indent:30px; text-align:justify">Теперь построим диаграмму разброса (рассеяния), показывающую зависимость <strong>MPG</strong> (<strong>Miles Per Gallon</strong>, сколько миль машина может проехать на 1 галлоне топлива) от мощности двигателя в лошадиных силах (<strong>horsepower</strong>):</div>

```python
sns.relplot(data=cars, x="horsepower", y="mpg", hue="cylinders", size="displacement")
```

<div style="text-indent:30px; text-align:justify">Обратите внимание, что Seaborn сама выполнила разбиение данных на подмножества. С помощью параметра <strong>hue</strong> (оттенок) выделены цветом подмножества по количеству цилиндров в двигателях, а благодаря параметру  <strong>size</strong>  сформированы подмножества, обозначаемые размерами точек в зависимости от рабочего объема двигателей (литража). </div>
<div style="text-indent:30px; text-align:justify">А теперь представьте, сколько бы кода потребовало построение того же самого графика в Matplotlib?</div>
<div style="text-indent:30px; text-align:justify">Диаграмма разброса (также называемая  графиком рассеяния) – это точечная диаграмма, получаемая путем нанесения на график экспериментальных данных в виде точек. Координаты точек на графике соответствуют значениям показателя качества <strong>у</strong> и влияющего на него фактора <strong>х</strong>. Диаграмма разброса позволяет сформулировать и проверить гипотезу о наличии или отсутствии корреляционной связи между двумя признаками (характеристика качества и влияющий на нее фактор).</div>

<h2 style="text-align:center">Линейный график</h2>

<div style="text-indent:30px; text-align:justify">Функция  <strong>relplot()</strong> может также создать линейный график, если использовать параметр  <strong>kind="line"</strong> (по умолчанию параметр имеет значение <strong>scatter</strong>):</div>

```python
sns.relplot(data=cars, x="horsepower", y="displacement", kind="line")
```
