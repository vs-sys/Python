<h2 style="text-align:center">Функция <strong>imshow()</strong></h2>

<div style="text-indent:30px; text-align:justify">Функция <strong>imshow()</strong> интерпретирует массив как растровое изображение.</div>
<div style="text-indent:30px; text-align:justify">Изображение будет иметь один квадрат для каждого элемента массива. Цвет каждого квадрата определяется значением соответствующего элемента массива и цветовой картой, используемой функцией <strong>imshow()</strong>.</div>

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 10)  # Функция генерирует равномерное распределение чисел в интервале [0, 1]
plt.imshow(data, cmap = "jet") # Создаем растровое изображение

plt.show() # Отображение диаграммы
```

<div style="text-indent:30px; text-align:justify">Еще один пример:</div>

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

# Наборы данных
r = np.arange(-4, 4, 0.25)
x, y = np.meshgrid(r, r)          # Создаем двумерный массив (сетку координат)
z = np.sin(np.sqrt(x**2 + y**2))  # Функция двух переменных

plt.imshow(z) # Создаем растровое изображение
plt.show()    # Отображение диаграммы
```

<div style="text-indent:30px; text-align:justify">Функция <strong>meshgrid</strong> применяется для формирования двумерной сетки координат <strong>x</strong> и <strong>y</strong> на основе заданных числовых диапазонов, что упрощает построение трехмерных графиков и выполнение операций с матрицами. Она позволяет трансформировать одномерные массивы в двумерные матрицы, создавая универсальную сетку для пар координат <strong>x</strong> и <strong>y</strong>, необходимую для ряда функций.</div>


<div style="text-indent:30px; text-align:justify; margin-top:10px">Можно также загрузить картинку из файла при помощи функции <strong>imread</strong>. Это будет трехмерный массив <strong>numpy.array</strong>, размерностью <strong>Ширина x Высота x 3</strong>, который описывает RGB-координаты всех пикселей изображения. Такой массив состоит из трех срезов, которые задают RGB-координаты цветов на изображении:</div>
<div style="text-indent:30px; text-align:justify; margin-top:10px"><strong>a[ : , : , 0]</strong></div>
<div style="text-indent:30px; text-align:justify; margin-top:10px"><strong>a[ : , : , 1]</strong></div>
<div style="text-indent:30px; text-align:justify; margin-top:10px"><strong>a[ : , : , 2]</strong></div> 

```python
# Импорт библиотек
import matplotlib.pyplot as plt
import numpy as np

picture = plt.imread('Python.jpg')    # Загружаем картинку в массив
print(type(picture), picture.shape)   # Размеры массива picture

plt.imshow(picture)                   # Вывод изображения
plt.axis('off')                       # Отключить оси
plt.show()                            # Отображение диаграммы
```

<div style="text-indent:30px; text-align:justify">Если мы установим элементы массива, задающие зеленые и синие координаты цветов RGB, равными 0, то получим красную часть изображения:</div>

```python
picture = plt.imread('Python.jpg')    # Загружаем картинку в массив

red = picture.copy()  # Делаем копию массива
red[:, :,[1,2]] = 0   # Обнуляем слои G(reen) и B(lue)

plt.imshow(red)       # Вывод изображения
plt.axis('off')                       # Отключить оси
plt.show()            # Отображение диаграммы
```

<div style="text-indent:30px; text-align:justify">Мы можем обрезать изображение, взяв часть массива:</div>

```python
plt.imshow(picture[0:300, 300:650, :])
plt.axis('off')       # Отключить оси  
plt.show()            # Отображение диаграммы
```

<h2 style="text-align:center">Построение 3D-графиков</h2>
<div style="text-indent:30px; text-align:justify">До этого момента все графики, которые мы строили, были двумерные, а Matplotlib позволяет строить также 3D-графики. </div>

<h3 style="text-align:center">Трехмерные графики</h3>

<div style="text-indent:30px; text-align:justify">Задается параметрически: <strong>x=f1(t)</strong>, <strong>y=f2(t)</strong>, <strong>z=f3(t)</strong>.</div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

# Набор данных
t = np.linspace(0, 4 * np.pi, 100)
x = np.cos(t)
y = np.sin(t)
z = t / (4 * np.pi)

plt.subplot(projection='3d')   # Создать 3D область

plt.plot(x, y, z, linewidth=5) # Линейный график

plt.show()                     # Отображение графика
```

<h3 style="text-align:center">Точечный график (диаграмма рассеяния)</h3>

<div style="text-indent:30px; text-align:justify">Для построения диаграммы рассеяния используется функция <strong>scatter()</strong>:</div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

# Наборы данных
np.random.seed(123)
xs1 = np.random.randint(30,40,100)
ys1 = np.random.randint(20,30,100)
zs1 = np.random.randint(10,20,100)

xs2 = np.random.randint(50,60,100)
ys2 = np.random.randint(30,40,100)
zs2 = np.random.randint(50,70,100)

xs3 = np.random.randint(10,30,100)
ys3 = np.random.randint(40,50,100)
zs3 = np.random.randint(40,50,100)

fig = plt.figure()
ax = fig.add_subplot(projection='3d') # Создать 3D область

ax.scatter(xs1, ys1, zs1)                    # Точечный график
ax.scatter(xs2, ys2, zs2, c='r', marker='^') # Точечный график
ax.scatter(xs3, ys3, zs3, c='g',marker='*')  # Точечный график

plt.show() # Отображение графика
```

<div style="text-indent:30px; text-align:justify">К сожалению, <i>inline</i> трехмерную картинку нельзя вертеть мышкой. Это можно делать с трехмерными картинками в отдельных окнах.</div>
<div style="text-indent:30px; text-align:justify">Но можно задать, с какой стороны мы смотрим. Функция <strong>view_unit()</strong> подстраивает точку обзора, откуда можно будет рассмотреть поверхность, изменяя аргументы <strong>elev</strong> и <strong>azim</strong>. С помощью их комбинирования можно получить картинку, изображенную с любого угла. Первый аргумент настраивает высоту, а второй — угол поворота поверхности.</div>

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d') # Создать 3D область

ax.view_init(elev=20,azim=45)

ax.scatter(xs1, ys1, zs1)                    # Точечный график
ax.scatter(xs2, ys2, zs2, c='r', marker='^') # Точечный график
ax.scatter(xs3, ys3, zs3, c='g',marker='*')  # Точечный график

plt.show() # Отображение графика
```

<h2 style="text-align:center">Поверхности</h2>

<h3 style="text-align:center">Каркасная поверхность</h3>

<div style="text-indent:30px; text-align:justify">Для построения каркасной поверхности используется функция  <strong>plot_wireframe()</strong>:</div>

```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt

x = y = np.linspace(-3, 3, 30)   # x, y - одномерные массивы
X, Y = np.meshgrid(x, y)         # X, Y - двумерные массивы
Z = np.sin(np.sqrt(X**2 + Y**2)) # Z - двумерный массив

fig = plt.figure()
ax = fig.add_subplot(projection='3d') # Создать 3D область

ax.plot_wireframe(X, Y, Z, color="blue") # Построение каркасной поверхности
ax.set_title("Каркас")

plt.show() # Отображение диаграммы
```

<div style="text-indent:30px; text-align:justify">Функция <strong>meshgrid()</strong> из модуля <strong>numpy</strong> позволят получить матрицу координат (координатную сетку) из координатных векторов (одномерные массивы <strong>x</strong> и <strong>y</strong>). Если, например, у нас есть два одномерных вектора координат, то передав их в качестве аргументов в  <strong>meshgrid()</strong> мы получим две матрицы (двумерные массивы <strong>X</strong> и <strong>Y</strong>), в которых элементы будут составлять пары, заполняя все пространство, определяемое этими векторами.</div>

<h3 style="text-align:center">Окрашенная поверхность</h3>
<div style="text-indent:30px; text-align:justify">Для построения окрашенной поверхности используется функция  <strong>plot_surface()</strong>:</div>

```python
# Импорт модулей
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()                    # Создать рисунок
ax = fig.add_subplot(projection='3d') # Добавить 3D область

x = y = np.linspace(-10, 10, 30)        # x, y - одномерные массивы
X, Y = np.meshgrid(x, y)              # X, Y - двумерные массивы

#Z1 = np.zeros((len(Y), len(X)))       # Z1=0 - плоскость
Z2 = np.sin(X) * np.sin(Y) / (X * Y)

#ax.plot_surface(X, Y, Z1)             # Построение поверхности
ax.plot_surface(X, Y, Z2, cmap='jet') # Построение поверхности

ax.elev = 40                          # Повернем немного на себя

plt.show()                            # Отображение диаграмм
```

<div style="text-indent:30px; text-align:justify">Построим бублик — параметрическую поверхность с параметрами <strong>θ</strong>
, <strong>φ</strong></div>

```python
# Импорт модулей
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()                    # Создать рисунок
ax = fig.add_subplot(projection='3d') # Добавить 3D область

t = np.linspace(0, 2 * np.pi, 50)     # t - одномерный массив
th, ph = np.meshgrid(t, t)            # th, ph - двумерные массивы

r = 0.5

x, y, z = (1 + r * np.cos(ph)) * np.cos(th), (1 + r * np.cos(ph)) * np.sin(th), r * np.sin(ph)

ax.elev = 60                          # Повернем немного на себя

ax.plot_surface(x, y, z, cmap="brg")  # Построение поверхности

plt.show()                            # Отображение диаграмы
```

<div style="text-indent:30px; text-align:justify">Следующий пример использует класс <strong>LinearSegmentedColormap</strong>, производный от <strong>Colormap</strong>, чтобы создать градиент перехода от красного цвета к синему через фиолетовый.</div>

```python
# Импорт модулей
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

fig = plt.figure()                    # Создать рисунок
ax = fig.add_subplot(projection='3d') # Добавить 3D область

t = np.linspace(0, 2 * np.pi, 50)     # t - одномерный массив
th, ph = np.meshgrid(t, t)            # th, ph - двумерные массивы

r = 0.5

x, y, z = (1 + r * np.cos(ph)) * np.cos(th), (1 + r * np.cos(ph)) * np.sin(th), r * np.sin(ph)

ax.elev = 60                          # Повернем немного на себя

cmap = LinearSegmentedColormap.from_list ('red_blue', ['r', 'c', 'b'], 256)
ax.plot_surface(x, y, z, cmap=cmap)  # Построение поверхности

plt.show()                            # Отображение диаграмы
```

<div style="text-indent:30px; text-align:justify">Здесь для создания цветовой карты используется статический метод <strong>from_list()</strong>, который принимает три параметра:</div>
<div style=" margin-left:50px">
<li style="text-align:justify">Имя создаваемой карты.</li>
<li style="text-align:justify">Список цветов, начиная с цвета для минимального значения на графике (красный - <strong>'r'</strong>), через промежуточные цвета (у нас это циан - <strong>'c'</strong>) к цвету для максимального значения функции (голубой - <strong>'b'</strong>).</li>
<li style="text-align:justify">Количество цветовых переходов. Чем это число больше, тем более плавный градиент, но тем больше памяти он занимает.</li>
</div>
